```python
import pandas as pd
import matplotlib.pyplot as plt
#3.10.7
```

```python
MC = pd.read_csv('List of Countries by number of Internet Users - Sheet1.csv')
MC.head()
```

```python
grouped_by_numbers= MC.groupby(["Population"]).count()
grouped_by_numbers
```

```python
grouped_by_numbers.plot(x="Country or Area", y="Rank", kind="pie")
```
